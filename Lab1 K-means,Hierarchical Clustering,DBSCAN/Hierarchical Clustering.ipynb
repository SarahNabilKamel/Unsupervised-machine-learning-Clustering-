# **üå≥ Hierarchical Clustering**

Hierarchical clustering is an **unsupervised learning** technique used to cluster data into groups **without needing to specify the number of clusters** beforehand. <br>
It builds a hierarchical structure of clusters using either a **bottom-up (Agglomerative) üîΩ** or **top-down (Divisive) üîº** approach.

<br>

#### **üìå Types of Hierarchical Clustering**

1. **Agglomerative Clustering üîó:**  
   - Each data point starts as its own cluster.  
   - The closest clusters are merged iteratively until a single large cluster is formed.  

2. **Divisive Clustering ‚úÇÔ∏è:**  
   - Starts with all data points in one cluster.  
   - The cluster is recursively split until each data point is its own cluster.  

<br>

---

<br>

#### **üîÑ Steps of Hierarchical Clustering**

Hierarchical clustering follows a series of steps to form a hierarchical structure of clusters. 
<br>Below is a detailed breakdown of the process:

<br>

##### **1Ô∏è‚É£ Compute Distance Matrix üìè**
- Calculate the pairwise distances between all data points.
- The choice of distance metric (**Euclidean, Manhattan, Cosine, etc.**) impacts the clustering outcome.
<br>

##### **2Ô∏è‚É£ Initialize Each Data Point as a Cluster üîπ**
- Each data point is treated as an individual cluster.
- At the beginning, if there are **N** data points, there are **N** clusters.

<br>

##### **3Ô∏è‚É£ Merge Closest Clusters üîó**
- Identify the two clusters that are closest to each other.
- Merge them into a single cluster.
- The way clusters are merged depends on the **linkage method**:

| Linkage Type  | How Distance is Measured üìè |
|--------------|---------------------------|
| **Single Linkage** | Distance between closest points |
| **Complete Linkage** | Distance between farthest points |
| **Average Linkage** | Average distance between all points |
| **Centroid Linkage** | Distance between cluster centroids |

<br>

##### **4Ô∏è‚É£ Update Distance Matrix üîÑ**
- After merging, update the distance matrix to reflect the new cluster.
- Recalculate distances between the newly formed cluster and remaining clusters.

<br>

##### **5Ô∏è‚É£ Repeat Until One Cluster Remains üîÅ**
- Continue merging the closest clusters iteratively.
- The process stops when **all points belong to a single cluster**.

<br>

##### **6Ô∏è‚É£ Build the Dendrogram üå≤**
- The hierarchical structure is visualized using a **dendrogram**.
- The **height** of the dendrogram represents the **distance** at which clusters were merged.
- Cutting the dendrogram at a chosen height determines the number of clusters.

<br>

##### **7Ô∏è‚É£ Determine Final Clusters ‚úÇÔ∏è**
- Decide at which level to cut the dendrogram.
- This determines the final number of clusters.

<br>

##### **‚úÖ Summary**
1. Compute the distance matrix between points.
2. Treat each data point as an individual cluster.
3. Merge the closest clusters based on a chosen linkage method.
4. Update the distance matrix after each merge.
5. Repeat until all points form a single cluster.
6. Visualize the hierarchical structure using a dendrogram.
7. Cut the dendrogram at the desired level to obtain clusters.

[Whether you want Agglomerative Clustering or Divisive Clustering, the steps are the same, but the direction of implementation will be different. üòä]

- When working with **small to medium-sized datasets üìÇ**.  
- When analyzing the **structural relationships** in data üîç.  
- When the number of clusters is **unknown ‚ùì**.  
- When an **interpretable clustering process** is needed üßê.

Hierarchical clustering is a **powerful üí™** technique for data analysis, but choosing the **right parameters** is essential to achieve accurate results. Modern techniques can enhance its performance, making it feasible for larger datasets.

#### **üìä Types of Data Suitable for Hierarchical Clustering**

Hierarchical clustering is ideal for certain types of data that have a **natural hierarchical structure**.
<br>Below are the data types that work well with this method:

<br>

##### **1Ô∏è‚É£ Numerical Data üî¢**  
Hierarchical clustering efficiently handles continuous numerical data, such as:  
##### ‚úÖ **Advantage:**  
‚úî Performs well with numerical data when using distance metrics like **Euclidean Distance** or **Manhattan Distance**.  

<br>

##### **2Ô∏è‚É£ Categorical Data üî§**  
It can be applied to categorical data (**e.g., gender, color, age group**), but requires conversion into numerical format using:  
- **Hamming Distance** to measure differences between categorical values.  
- **One-Hot Encoding** to transform categories into numerical vectors.  

##### ‚ùå **Challenge:**  
‚úñ Less effective for pure categorical data compared to methods like **K-Modes**.  


##### **3Ô∏è‚É£ Text Data üìÑ**  
To apply hierarchical clustering to text, it must first be transformed into numerical representations using:  
- **TF-IDF (Term Frequency-Inverse Document Frequency).**  
- **Word Embeddings (e.g., Word2Vec, GloVe).**  
- **Cosine Similarity** instead of Euclidean distances.  

##### ‚úÖ **Advantage:**  
‚úî Useful for **document clustering, topic analysis, and text classification**.  

##### ‚ùå **Challenge:**  
‚úñ Requires extensive preprocessing due to the non-numeric nature of text.  

<br>

##### **4Ô∏è‚É£ Image Data üñºÔ∏è**  
Hierarchical clustering is used in:  
- **Classifying images based on visual similarity.**  
- **Grouping similar medical images or facial patterns.**  
- **Finding similar images in large datasets.**  

##### ‚úÖ **Advantage:**  
‚úî Works well with **feature representations extracted from deep neural networks**.  

##### ‚ùå **Challenge:**  
‚úñ Requires **dimensionality reduction (e.g., PCA, Autoencoders)** for better efficiency.  

<br>

##### **5Ô∏è‚É£ Time-Series Data ‚è≥**  
Used in:  
- **Weather pattern clustering.**  
- **Stock market trend analysis.**  
- **Gait pattern classification in sports science.**  

##### ‚úÖ **Advantage:**  
‚úî Can analyze time-series data after computing similarity using **Dynamic Time Warping (DTW)**.  

##### ‚ùå **Challenge:**  
‚úñ **Computationally expensive** for long and complex time-series data.  

<br>

---

#### **‚ùå When is Hierarchical Clustering Not Suitable?**
- **Big Data (Massive Datasets) ‚ö°:** Due to its **O(n¬≥) complexity**, it struggles with millions or billions of data points without optimization.  
- **Non-Hierarchical Data Structures üèóÔ∏è:** If no **clear hierarchical pattern** exists, alternative algorithms like **K-Means or DBSCAN** may perform better.  

---

<br>

#### **üìå Conclusion**
‚úî **Best for:** Numerical, categorical, text, image, and time-series data.  
‚ùå **Not ideal for:** Massive datasets or data without a hierarchical structure.

#### **üìå Can Hierarchical Clustering Handle High-Dimensional Data?**  

Hierarchical clustering **can** work with high-dimensional data, but it faces **significant challenges** and requires **optimization techniques** to improve performance. Here are the key details:  

<br>

##### **üöß Challenges of High-Dimensional Data**
##### **1Ô∏è‚É£ Curse of Dimensionality**
- When the number of features is very high, distance calculations become **less reliable** because all points appear equally far apart.
- This issue leads to **degraded clustering performance**.

##### **2Ô∏è‚É£ High Computational Complexity**
- Hierarchical clustering has a **O(n¬≥) complexity**, making it extremely slow for large datasets.
- As dimensions increase, **memory and time consumption grow exponentially**.

##### **3Ô∏è‚É£ Choosing the Right Distance Metric**
- **Euclidean Distance** is not ideal for high-dimensional data since it becomes less meaningful.
- **Solution:** Use alternative distance measures such as:
  - **Cosine Similarity** (suitable for text and vector representations).
  - **Mahalanobis Distance** (accounts for feature correlations).

<br>

---

#### **üöÄ How to Improve Hierarchical Clustering for High-Dimensional Data?**  

##### **1Ô∏è‚É£ Dimensionality Reduction (PCA or t-SNE)**
- **Principal Component Analysis (PCA):**  
  - Reduces the number of features while preserving most of the information.  
  - Makes clustering more accurate and computationally efficient.
  
- **t-SNE or UMAP:**  
  - Non-linear techniques useful for **complex data visualization and clustering**.

‚úÖ **Outcome:** Reduces features from hundreds to **10-50 dimensions**, improving clustering performance.

---

##### **2Ô∏è‚É£ Hybrid Clustering Approach**
- **Instead of applying hierarchical clustering directly, combine it with K-Means:**
  1. **Use K-Means to reduce the number of data points** (e.g., cluster data into 100 groups).
  2. **Apply hierarchical clustering on K-Means centroids** for a structured hierarchy.
  
‚úÖ **Outcome:** Significantly reduces computational cost without sacrificing accuracy.

---

##### **3Ô∏è‚É£ Use Data Sampling Instead of Full Dataset**
- For **millions of data points**, apply hierarchical clustering on a **representative subset** (e.g., 10-20% of the dataset).
- Classify remaining data points using statistical techniques.

‚úÖ **Outcome:** Faster clustering while maintaining high quality.

---

##### **üìå Conclusion**
‚úî **Can hierarchical clustering handle high-dimensional data?**  
‚úÖ Yes, but **it becomes slow and inefficient** unless optimized using:
  - **PCA or t-SNE for dimensionality reduction.**
  - **Alternative distance metrics like Cosine Similarity.**
  - **Hybrid approaches like combining it with K-Means or using data sampling.**

‚ùå **For extremely large datasets, K-Means or DBSCAN might be better alternatives.**

#### **üìå How to Measure the Effectiveness of Hierarchical Clustering?**  
Since hierarchical clustering is **unsupervised learning**, evaluating its quality is not as straightforward as classification, where predefined labels exist. However, several techniques can help assess its effectiveness:

---

##### **üîπ 1. Evaluating Cluster Cohesion & Separation**  
A good clustering result should have **high intra-cluster similarity** and **low inter-cluster similarity**.

##### **üìè A. Silhouette Score**
- Measures how similar a point is to its own cluster compared to other clusters.
- Values range between **-1 and +1**:
  - **+1:** Well-clustered point.
  - **0:** Point on the cluster boundary.
  - **-1:** Misclassified point.

‚úÖ **Advantage:** Useful when the ideal number of clusters is unknown.  
üîß **Formula:**  



- **a:** Mean distance between a point and others in the same cluster.
- **b:** Mean distance between a point and the nearest other cluster.

<br>

##### **üìè B. Davies-Bouldin Index**
- Measures cluster overlap (lower values indicate better clustering).
- Optimal value: Close to zero.

```python 
from sklearn.metrics import davies_bouldin_score

db_score = davies_bouldin_score(X, labels)
print(f'Davies-Bouldin Index: {db_score}')
```
<br>

##### **üìè C. Dunn Index**
- Measures the ratio between the shortest inter-cluster distance and the longest intra-cluster distance.
- Higher values indicate better clustering.

<br>

---

##### **üîπ 2. Dendrogram Analysis**
A dendrogram helps visualize how data points are merged into clusters.
The optimal number of clusters can be determined by cutting the dendrogram at a specific level. <br>
‚úÖ Advantage: Provides a visual representation of hierarchical clustering.

---

##### **üîπ 3. Comparison with Other Clustering Algorithms**
If you're unsure whether hierarchical clustering is the best choice, compare it with:
- K-Means (when a fixed number of clusters is known).
- DBSCAN (for irregular or noisy data).
‚úÖ Advantage: Helps choose the best clustering method based on quality.

---

##### **üîπ 4. Cluster Stability**
If you modify the dataset slightly, do clusters remain the same?
‚úî Good clustering should be stable and not change significantly when adding new data.

üõ† How to test:

- Take a subset of the data, apply clustering, then take another subset and compare results.
- If clusters change significantly, the model may not be reliable.

````python
# üéØ Silhouette Score Calculation and Evaluation

for k in range(2, 10): 
    clustering = AgglomerativeClustering(n_clusters=k, linkage='ward')
    labels = clustering.fit_predict(df)
    score = silhouette_score(df, labels)
    print(f"n_clusters = {k}, Silhouette Score = {score:.3f}")
````

````python
for method in ['ward', 'complete', 'average', 'single']:
    clustering = AgglomerativeClustering(n_clusters=2, linkage=method)
    labels = clustering.fit_predict(df)
    score = silhouette_score(df, labels)
    print(f"Linkage: {method}, Silhouette Score: {score:.3f}")

    pca = PCA(n_components=2)
    X_pca = pca.fit_transform(df)

    plt.figure(figsize=(8, 5))
    sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=labels, palette="Set1", s=80, edgecolor='k')
    plt.title(f'Linkage: {method}, Silhouette Score: {score:.3f}')
    plt.xlabel("PCA Component 1")
    plt.ylabel("PCA Component 2")
    plt.legend(title="Clusters")
    plt.show()  

# üü¢ Drawing a dendrogram
    plt.figure(figsize=(10, 5))
    linked = linkage(df, method='ward')
    dendrogram(linked, truncate_mode='level', p=5)  # Show only the first 5 levels of the tree.
    plt.title("Dendrogram for Hierarchical Clustering")
    plt.xlabel("Sampled Data Points")
    plt.ylabel("Distance")
    plt.show()
````

### **‚úÖ Advantages of Hierarchical Clustering**

1. **No need to predefine the number of clusters üî¢**  
   - Unlike K-Means, hierarchical clustering does not require specifying the number of clusters (**K**) beforehand.  

2. **Easy to interpret and understand üëÄ**  
   - The **dendrogram üå≤** visually represents how data points are related.  

3. **Flexibility in choosing the clustering level üéöÔ∏è**  
   - The dendrogram can be cut at different levels to obtain a specific number of clusters.  

4. **Works well with hierarchical data üìä**  
   - If the data has an inherent hierarchical structure, this method provides accurate clustering.  

5. **Utilizes full similarity information üìè**  
   - Uses a complete distance matrix, making it more precise than K-Means.  

<br>

### **‚ö†Ô∏è Disadvantages of Hierarchical Clustering**

1. **High computational complexity üñ•Ô∏è‚è≥**  
   - The algorithm has a complexity of **O(n¬≥)**, making it impractical for large datasets.  

2. **Irreversible merging process üîÑ‚ùå**  
   - Once two clusters are merged or split, they cannot be undone, which can lead to permanent errors.  

3. **Sensitive to noise and outliers ‚ö°üìâ**  
   - Anomalous data points may significantly impact clustering results.  

4. **Depends on the chosen distance metric üìê**  
   - Results can vary depending on the distance calculation method (**Euclidean, Manhattan, Cosine, etc.**).  

5. **Not suitable for very large datasets üèóÔ∏è**  
   - Storing and processing a full distance matrix can be memory-intensive.  

<br>

### **üöÄ Advanced Insights on Hierarchical Clustering**

1. **Choosing the right linkage method is crucial üéØ**  
   - Different linkage methods produce different results:  
     - **Single Linkage üîó:** Merges clusters based on the closest pair of points.  
     - **Complete Linkage üîÑ:** Merges clusters based on the farthest pair of points.  
     - **Average Linkage üìä:** Uses the average distance between all points in two clusters.  
     - **Centroid Linkage üéØ:** Merges clusters based on their centroid positions.  

2. **Performance can be improved with modern techniques ‚ö°**  
   - Methods like **HDBSCAN (Hierarchical Density-Based Clustering) üß†** enhance efficiency and accuracy.  

3. **Careful dendrogram analysis is required üî¨**  
   - Cutting the tree at the wrong level can lead to incorrect clusters.  

4. **Can be combined with other clustering algorithms üîÑ**  
   - Applying **K-Means üéØ** after **Hierarchical Clustering** can refine the results.  

5. **Approximations reduce complexity ‚è©**  
   - Algorithms like **BIRCH üçÉ** and **CURE üõ†Ô∏è** approximate the clustering process, reducing computational costs.
